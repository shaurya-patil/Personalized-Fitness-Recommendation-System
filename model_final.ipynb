```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_absolute_error

final_data_path = "dataset/final_data.csv"
fitness_data_path = "dataset/fitness_data.csv"
final_df = pd.read_csv(final_data_path)
fitness_df = pd.read_csv(fitness_data_path)

final_df = final_df.drop(columns=['Unnamed: 0'], errors='ignore')

fitness_df.rename(columns={'Current Weight (kg)': 'Weight (kg)'}, inplace=True)

X = fitness_df[['Body Fat %', 'Weight (kg)', 'Gender']]
y = fitness_df['Time to Target (wks)']

imputer = SimpleImputer(strategy='mean')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

y_imputed = imputer.fit_transform(y.values.reshape(-1, 1)).flatten()

X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)

mae = mean_absolute_error(y_test, predictions)
print(f"Mean Absolute Error: {mae}")
```

```python
fitness_df
```

```python
importances = model.feature_importances_
feature_names = X_train.columns
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align="center")
plt.xticks(range(X_train.shape[1]), feature_names[indices], rotation=45)
plt.xlim([-1, X_train.shape[1]])
plt.ylabel("Importance")
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predictions vs Actual Values")
plt.show()

errors = y_test - predictions

plt.figure(figsize=(10, 6))
sns.histplot(errors, bins=30, kde=True)
plt.title("Distribution of Prediction Errors")
plt.xlabel("Error (Actual - Predicted)")
plt.ylabel("Frequency")
plt.show()

X = fitness_df.drop('Client Name', axis=1)
correlation_matrix = X.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
def detect_critical_disease(row):
    bp_cols = ["Systolic BP1", "Systolic BP2", "Systolic BP3", "Systolic BP4"]
    avg_bp = row[bp_cols].mean(skipna=True)

    if avg_bp > 130 or row.get("Cholesterol", 0) > 200:
        return "Cardiovascular Disease"

    if row.get("Total Sugar Intake (grams)", 0) > 50:
        return "Diabetes"

    bmi = row.get("Body Mass Index (BMI)", None)
    if bmi is not None and (bmi > 30):
        return "Obese"

    if row.get("Exposure to Secondhand Smoke - Home") == "Yes" or row.get("Source of Air Pollution Exposure") == "Yes":
        return "Lung Issues"

    if row.get("Depression Scale Score", 0) > 10 or row.get("Anxiety Scale Score", 0) > 10:
        return "Mental Health Issues"

    if row.get("Sleep Duration Per Night (Hours)", 10) < 5 or row.get("Sleep Quality Rating", 5) < 3:
        return "Sleep Disorder"

    return "None"

df2_merged["Disease"] = df2_merged.apply(detect_critical_disease, axis=1)
```

```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.multioutput import MultiOutputRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

numerical_target_columns = [
    'Daily Calorie Target', 'Protein', 'Sugar', 'Sodium', 'Calories',
    'Carbohydrates', 'Fiber', 'Fat', 'Breakfast Calories',
    'Breakfast Protein', 'Breakfast Carbohydrates', 'Breakfast Fats',
    'Lunch Calories', 'Lunch Protein', 'Lunch Carbohydrates', 'Lunch Fats',
    'Dinner Calories', 'Dinner Protein.1', 'Dinner Carbohydrates.1', 'Dinner Fats',
    'Snacks Calories', 'Snacks Protein', 'Snacks Carbohydrates', 'Snacks Fats'
]

text_columns = ['Breakfast Suggestion', 'Lunch Suggestion', 'Dinner Suggestion', 'Snack Suggestion']

model = SentenceTransformer('all-MiniLM-L6-v2')

df['Body Mass Index (BMI)'] = df['Weight'] / ((df['Height']/100) ** 2)
trainer['Body Mass Index (BMI)'] = trainer['Body Weight (kg)'] / ((trainer['Body Height (cm)']/100) ** 2)

df_diseases = df['Disease'].unique().tolist()
trainer_diseases = trainer['Disease'].unique().tolist()

df_diseases_for_embedding = [d for d in df_diseases if d]
trainer_diseases_for_embedding = [d for d in trainer_diseases if d]

df_embeddings = model.encode(df_diseases_for_embedding)
trainer_embeddings = model.encode(trainer_diseases_for_embedding)

similarity_matrix = cosine_similarity(df_embeddings, trainer_embeddings)

disease_mapping = {}
for i, df_disease in enumerate(df_diseases_for_embedding):
    most_similar_idx = np.argmax(similarity_matrix[i])
    most_similar_disease = trainer_diseases_for_embedding[most_similar_idx]
    similarity_score = similarity_matrix[i][most_similar_idx]
    disease_mapping[df_disease] = {
        'mapped_to': most_similar_disease,
        'similarity': similarity_score
    }

if '' in df_diseases:
    if '' in trainer_diseases:
        disease_mapping[''] = {'mapped_to': 'None', 'similarity': 1.0}
    else:
        disease_mapping[''] = {'mapped_to': trainer_diseases[0], 'similarity': 0.0}
        print(f"Empty disease mapped to '{trainer_diseases[0]}' as default")

if '' in trainer_diseases and '' not in df_diseases:
    print("Adding a dummy row with empty disease to training data")
    dummy_row = df.iloc[0].copy()
    dummy_row['Disease'] = 'None'
    for col in numerical_target_columns:
        if col in df.columns:
            dummy_row[col] = df[col].mean()
    df = pd.concat([df, pd.DataFrame([dummy_row])], ignore_index=True)

df['Disease'] = df['Disease'].map(lambda x: disease_mapping.get(x, {'mapped_to': x})['mapped_to'])

# Handle NaN values in the dataset
for col in numerical_target_columns:
    if col in df.columns and df[col].isna().any():
        df[col] = df[col].fillna(df[col].mean())

for col in text_columns:
    if col in df.columns and df[col].isna().any():
        most_common = df[col].mode()[0]
        df[col] = df[col].fillna(most_common)

all_diseases = list(set(df['Disease'].unique()) | set(trainer['Disease'].unique()))

label_encoder = LabelEncoder()
label_encoder.fit(all_diseases)

df['Disease_encoded'] = label_encoder.transform(df['Disease'])

X = df[['Body Mass Index (BMI)', 'Disease_encoded']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y_numerical = df[numerical_target_columns]

text_encoders = {}
encoded_text_targets = {}
for col in text_columns:
    text_encoders[col] = LabelEncoder()
    encoded_text_targets[col] = text_encoders[col].fit_transform(df[col])

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_numerical, test_size=0.2, random_state=42)
numerical_model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
numerical_model.fit(X_train, y_train)
score = numerical_model.score(X_test, y_test)
print(f"Numerical model R² score: {score:.4f}")

text_models = {}
for col in text_columns:
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, encoded_text_targets[col], test_size=0.2, random_state=42
    )
    text_models[col] = RandomForestClassifier(n_estimators=100, random_state=42)
    text_models[col].fit(X_train, y_train)
    score = text_models[col].score(X_test, y_test)
    print(f"Text model for {col} accuracy: {score:.4f}")

trainer['Disease_encoded'] = label_encoder.transform(trainer['Disease'])
X_trainer = trainer[['Body Mass Index (BMI)', 'Disease_encoded']]
X_trainer_scaled = scaler.transform(X_trainer)

numerical_predictions = numerical_model.predict(X_trainer_scaled)
text_predictions = {}
for col in text_columns:
    encoded_predictions = text_models[col].predict(X_trainer_scaled)
    text_predictions[col] = text_encoders[col].inverse_transform(encoded_predictions)

merged_trainer = trainer.copy()

for i, col in enumerate(numerical_target_columns):
    merged_trainer[col] = numerical_predictions[:, i]

for col in text_columns:
    merged_trainer[col] = text_predictions[col]

for col in numerical_target_columns:
    if "Calories" in col:
        merged_trainer[col] = merged_trainer[col].round(0)
    else:
        merged_trainer[col] = merged_trainer[col].round(1)

print("Predictions completed and added to merged_trainer!")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load your data
df = pd.read_csv("dataset/merged_trainer.csv")

# Define columns
input_cols = ['Body Mass Index (BMI)', 'Body Weight (kg)', 'Predicted Target Weight (kg)',
              'Body Height (cm)', 'Predicted Time to Target (wks)', 'Disease']

num_cols = ['Body Mass Index (BMI)', 'Body Weight (kg)', 'Predicted Target Weight (kg)',
            'Body Height (cm)', 'Predicted Time to Target (wks)']

target_cols = ['Protein', 'Sugar', 'Sodium', 'Calories', 'Carbohydrates', 'Fiber', 'Fat',
               'Breakfast Calories', 'Breakfast Protein', 'Breakfast Carbohydrates', 'Breakfast Fats',
               'Lunch Calories', 'Lunch Protein', 'Lunch Carbohydrates', 'Lunch Fats',
               'Dinner Calories', 'Dinner Protein.1', 'Dinner Carbohydrates.1', 'Dinner Fats',
               'Snacks Calories', 'Snacks Protein', 'Snacks Carbohydrates', 'Snacks Fats']

# Drop missing target rows
df = df.dropna(subset=target_cols)

# Make copies
df_cat = df.copy()
df_lgb = df.copy()

# Handle NaN values in Disease column for both datasets
df_cat['Disease'] = df_cat['Disease'].fillna('None')
df_lgb['Disease'] = df_lgb['Disease'].fillna('None')

le = LabelEncoder()
df_lgb['Disease'] = le.fit_transform(df_lgb['Disease'])

df_cat['Disease'] = df_cat['Disease'].astype(str)

scaler = StandardScaler()
df_lgb[num_cols] = scaler.fit_transform(df_lgb[num_cols])
df_cat[num_cols] = scaler.transform(df_cat[num_cols])

X_cat = df_cat[input_cols]
X_lgb = df_lgb[input_cols]
y = df[target_cols]

# Train/test split
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_cat, y, test_size=0.2, random_state=42)
Xl_train, Xl_test, yl_train, yl_test = train_test_split(X_lgb, y, test_size=0.2, random_state=42)

# Train CatBoost
cat_model = MultiOutputRegressor(
    CatBoostRegressor(cat_features=[input_cols.index('Disease')], verbose=0)
)
cat_model.fit(Xc_train, yc_train)
yc_pred = cat_model.predict(Xc_test)

# Train LightGBM
lgb_model = MultiOutputRegressor(LGBMRegressor())
lgb_model.fit(Xl_train, yl_train)
yl_pred = lgb_model.predict(Xl_test)

# Evaluate
def evaluate(y_true, y_pred, label):
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f"{label} - MAE: {mae:.2f}, R2: {r2:.2f}")

print("\nEvaluation Metrics:\n")
evaluate(yc_test, yc_pred, "CatBoost")
evaluate(yl_test, yl_pred, "LightGBM")

import joblib
joblib.dump(cat_model, 'catboost_nutrition_model.pkl')
joblib.dump(lgb_model, 'lightgbm_nutrition_model.pkl')
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')

suggest_diet = suggest_diet.dropna(subset=['Disease'])
suggest_diet['Disease'] = suggest_diet['Disease'].astype(str)

suggest_embeddings = model.encode(suggest_diet['Disease'].tolist(), convert_to_tensor=True)

# Function to match a disease from df to the closest in suggest_diet
def match_disease_suggestion(disease):
    emb = model.encode(disease, convert_to_tensor=True)
    scores = util.pytorch_cos_sim(emb, suggest_embeddings)[0]
    best_idx = scores.argmax().item()
    return best_idx, scores[best_idx].item()

matched_rows = []

for disease in df['Disease']:
    if isinstance(disease, str) and disease.strip():
        idx, score = match_disease_suggestion(disease)
        suggestion_row = suggest_diet.iloc[idx]
        matched = {
            'Original Disease': disease,
            'Matched Disease': suggestion_row['Disease'],
            'Similarity Score': score
        }
        matched.update(suggestion_row.drop(labels=['Disease']).to_dict())
        matched_rows.append(matched)

final_result = pd.DataFrame(matched_rows)
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputRegressor
from sklearn.preprocessing import LabelEncoder, StandardScaler

class HealthChatBot:
    def __init__(self, data):
        self.raw_data = data.copy()
        self.nutrition_model = None
        self.time_model = None
        self.scaler = StandardScaler()
        self.label_encoder = LabelEncoder()

        self.input_cols_time = ['Body Mass Index (BMI)', 'Body Weight (kg)',
                                'Predicted Target Weight (kg)', 'Body Height (cm)', 'Disease']
        
        self.input_cols_nutrition = ['Body Mass Index (BMI)', 'Body Weight (kg)',
                                     'Predicted Target Weight (kg)', 'Body Height (cm)',
                                     'Predicted Time to Target (wks)', 'Disease']
        
        self.num_cols_time = ['Body Mass Index (BMI)', 'Body Weight (kg)',
                              'Predicted Target Weight (kg)', 'Body Height (cm)']
        
        self.num_cols_nutrition = ['Body Mass Index (BMI)', 'Body Weight (kg)',
                                   'Predicted Target Weight (kg)', 'Body Height (cm)',
                                   'Predicted Time to Target (wks)']
        
        self.target_cols = ['Protein', 'Sugar', 'Sodium', 'Calories', 'Carbohydrates', 'Fiber', 'Fat',
                            'Breakfast Calories', 'Breakfast Protein', 'Breakfast Carbohydrates', 'Breakfast Fats',
                            'Lunch Calories', 'Lunch Protein', 'Lunch Carbohydrates', 'Lunch Fats',
                            'Dinner Calories', 'Dinner Protein.1', 'Dinner Carbohydrates.1', 'Dinner Fats',
                            'Snacks Calories', 'Snacks Protein', 'Snacks Carbohydrates', 'Snacks Fats']
        
        self.suggestion_cols = ['Breakfast Suggestion', 'Lunch Suggestion', 'Dinner Suggestion', 'Snack Suggestion', 'suggestions']
        self._prepare_data()

    def _prepare_data(self):
        df = self.raw_data.dropna(subset=self.target_cols + ['Predicted Time to Target (wks)']).copy()
        df['Disease'] = df['Disease'].astype(str).fillna("None")
        df['Disease'] = self.label_encoder.fit_transform(df['Disease'])

        df[self.num_cols_nutrition] = self.scaler.fit_transform(df[self.num_cols_nutrition])

        time_df = df[self.input_cols_time + ['Predicted Time to Target (wks)']].copy()
        X_time = time_df[self.input_cols_time]
        y_time = time_df['Predicted Time to Target (wks)']
        self.time_model = RandomForestRegressor(n_estimators=100, random_state=42)
        self.time_model.fit(X_time, y_time)

        X_nutrition = df[self.input_cols_nutrition]
        y_nutrition = df[self.target_cols]
        self.nutrition_model = MultiOutputRegressor(RandomForestRegressor(n_estimators=100, random_state=42))
        self.nutrition_model.fit(X_nutrition, y_nutrition)

        self.suggestion_data = df[self.input_cols_nutrition + self.suggestion_cols].copy()

    def get_user_input(self):
        weight = float(input("Enter your weight (kg): "))
        height = float(input("Enter your height (cm): "))
        target_weight = float(input("Enter your target weight (kg): "))
        has_disease = input("Do you have any disease? (yes/no): ").lower().strip() == "yes"
        disease = input("What disease? (type or leave blank): ") if has_disease else "None"

        bmi = round(weight / ((height / 100) ** 2), 2)
        disease_encoded = self.label_encoder.transform([disease])[0] if disease in self.label_encoder.classes_ else 0

        input_time = pd.DataFrame([{
            'Body Mass Index (BMI)': bmi,
            'Body Weight (kg)': weight,
            'Predicted Target Weight (kg)': target_weight,
            'Body Height (cm)': height,
            'Disease': disease_encoded
        }])[self.input_cols_time]

        predicted_time = self.time_model.predict(input_time)[0]

        input_nutrition = input_time.copy()
        input_nutrition['Predicted Time to Target (wks)'] = predicted_time
        input_nutrition = input_nutrition[self.input_cols_nutrition]
        input_nutrition[self.num_cols_nutrition] = self.scaler.transform(input_nutrition[self.num_cols_nutrition])

        self.original_input = {
            'bmi': bmi,
            'weight': weight,
            'height': height,
            'target_weight': target_weight,
            'predicted_time': predicted_time,
            'disease': disease
        }
        return input_nutrition


    def predict_and_show(self, input_data):
        preds = self.nutrition_model.predict(input_data)[0]
        output = dict(zip(self.target_cols, preds))

        print("\n📋 Nutritional Recommendations:")
        print(f"Total Calories: {output['Calories']:.2f} kcal")
        print(f"Protein: {output['Protein']:.2f}g | Carbs: {output['Carbohydrates']:.2f}g | Fat: {output['Fat']:.2f}g")

        print("\n🕒 Predicted Time to Reach Target Weight:")
        print(f"{self.original_input['predicted_time']:.1f} weeks")

        print("\n🍽️ Meal-wise Suggestions (Calories | Protein | Carbs | Fats):")
        print(f"🥣 Breakfast: {output['Breakfast Calories']:.0f} kcal | {output['Breakfast Protein']:.1f}g | "
              f"{output['Breakfast Carbohydrates']:.1f}g | {output['Breakfast Fats']:.1f}g")
        print(f"🥗 Lunch: {output['Lunch Calories']:.0f} kcal | {output['Lunch Protein']:.1f}g | "
              f"{output['Lunch Carbohydrates']:.1f}g | {output['Lunch Fats']:.1f}g")
        print(f"🍛 Dinner: {output['Dinner Calories']:.0f} kcal | {output['Dinner Protein.1']:.1f}g | "
              f"{output['Dinner Carbohydrates.1']:.1f}g | {output['Dinner Fats']:.1f}g")
        print(f"🍪 Snacks: {output['Snacks Calories']:.0f} kcal | {output['Snacks Protein']:.1f}g | "
              f"{output['Snacks Carbohydrates']:.1f}g | {output['Snacks Fats']:.1f}g")

        # Closest suggestion
        print("\n📌 Based on your profile, here are meal suggestions:")
        closest = self.find_closest_suggestion(input_data.iloc[0])
        for col in self.suggestion_cols:
            if col in closest:
                print(f"{col.replace('Suggestion', '').strip()}: {closest[col]}")

    def find_closest_suggestion(self, user_row):
        df = self.suggestion_data.copy()
        df['distance'] = df[self.input_cols_nutrition].apply(lambda row: ((row - user_row) ** 2).sum(), axis=1)
        return df.sort_values(by="distance").iloc[0]

    def run(self):
        print("\n💬 Welcome to the Health Recommendation Chatbot!")
        input_data = self.get_user_input()
        self.predict_and_show(input_data)


df = pd.read_csv("dataset/FULL_DATASET.csv")
df.columns = df.columns.str.strip()

bot = HealthChatBot(df)
bot.run()
```
